```python
from prettytable import PrettyTable as pt
import os
from csv import *
from random import randint
class Restaurant:
    def user_or_admin(self):
        while True:
            print('   -----------------------------------')
            print(' |    |1| USER LOGIN |2| ADMIN  LOGIN  |')
            print('   -----------------------------------')
            user_or_admin=int(input(('press 1 for user login or register of user and  2 for admin login:')))
            while True:
                if(user_or_admin==1):
                    user_o.login_or_reg()
                elif(user_or_admin==2):
                    admin_o.admin_valadator()
                    admin_o.admin_option()
                else:
                    print('wrong input try again.....')
                    self.user_or_admin(self) 
class Admin(Restaurant):
    @staticmethod
    def admin_valadator():
        while True:
            name=input("enter admin id:")
            password=input("enter password:")
            if(name=='admin' and password=='1234'):
                break
            else:
                print('wrong user name or password try again............')
    def admin_option(self):
        while True:
            print('  ----------------------------------------------')
            print('|  |1|ADD ITEM |2|VIEW ITEMS |3| REMOVE AN ITEM  |')
            print('  ---------------------------------------------')
            self.choose=int(input(('press \n 1 for add item \n 2 for view items \n 3 for remove an item \n')))
            if(self.choose==1):
                admin_o.add_item()
            elif(self.choose==2):
                admin_o.view_item()
            elif(self.choose==3):
                admin_o.remove_item()    
                
    def add_item(self):
        with open('add_data.csv','a',newline='') as cf:
            c=writer(cf)
            while True:
                user_in=input('press \'yes\' to enter data and \'no\' to exit:')
                if(user_in.lower()=='yes'):
                    self.orderid=randint(1000,9999)
                    self.name=input('enter name:')
                    self.quantity=int(input('enter quntity:'))
                    self.price=int(input('enter price:'))
                    self.discount=int(input('enter discount:'))
                    self.stock=int(input('enter avilable stock:'))
                    l=[self.orderid,self.name,self.quantity,self.price,self.discount,self.stock]
                    c.writerow(l)
                    print("data sucessfully added")
                elif(user_in.lower()=='no'):
                    self.admin_option()
                            
    def view_item(self):
            with open('add_data.csv','r',newline='') as cf:
                c=reader(cf)
                flag=True
                for row in c:
                    if flag==True:
                        mytable=pt(row)
                        flag==False
                    else:
                        mytable.add_row(row)
                print(mytable)    
    def remove_item(self):
        print("before deleting")
        self.view_item()
        remove_id=input("enter remove id to remove an item:")
        l=[]
        with open('add_data.csv','r',newline='') as c:
            c=reader(c)
            for row in c:
                l.append(row)
        with  open('add_data.csv','w',newline='') as c:
            wo=writer(c)
            for i in l:  
                if(i[0]== remove_id):
                    continue
                else:
                    wo.writerow(i)            
        print("after deleting")
        self.view_item()                       
class user(Admin):
    def login_or_reg(self):
        print('   -----------------------------')
        print(' |    |1| LOGIN  |2| REGISTER    |')
        print('   -----------------------------')
        new_or_old_user=int(input('press  1 if you are already a user and 2 for user register:'))
        if(new_or_old_user==1):
            user_o.login()
        elif(new_or_old_user==2):
            self.fullname=input('enter name:')
            self.phoneno=int(input('enter phoneno:'))
            self.emailid=input('enter price emailid:')
            self.address=input('enter address:')
            self.password=input('enter password:')
            self.register(self.fullname,self.phoneno,self.emailid,self.address,self.password)
        else:
            print("wrong input try again")
    def register(self,fullname,phoneno,emailid,address,password):
        with open('reg.csv','a',newline='') as cf:
            c=writer(cf)  
            l1=[fullname,phoneno,emailid,address,password]
            c.writerow(l1)
            print("registered sucessfully ")
                
        self.login()   
    def login(self):
        print("you can login here")
        self.user_name=input('enter email id:')
        password=input('enter password:')
        with open('reg.csv','r',newline='') as cf:
            c=reader(cf)
            l=[]
            for row in c:
                l.append(row)
            for i in l: 
                if(i[2]== self.user_name and i[4]== password):
                    print('login sucesseded')
                    break
            else:
                print('wrong username or password ')
                self.login()
        self.main_user_page_board()
    def main_user_page_board(self):
        print('  -----------------------------------------------------------------')
        print('|  |1|PLACE ORDER |2|ORDER HISTORY |3| UPDATE DETAILS|  |0| LOGOUT  |')
        print('  -----------------------------------------------------------------')
        user_option=int(input('press \n 1 for place order \n 2 for order history \n 3 for update details \n 0 to logout'))
        if user_option==1:
            self.show_food_items()
            self.place_order()
        elif user_option==2:
            self.order_history()
        elif user_option==3:
            self.update_profile()
        elif user_option==0:
            super().user_or_admin()
            
             
    def show_food_items(self):
        with open('add_data.csv','r') as fo:
            ro=reader(fo)
            l=[]
            for i in ro:    
                l.append(i)
            count=1
            num=1
            for j in l:
                if count==1:
                    j[0]='s.no'  
                    j.remove(j[-2])
                    j.remove(j[-1])
                    mytable=pt(j)
                    count=0
                else:
                    j[0]= num
                    j.remove(j[-2])
                    j.remove(j[-1])
                    num+=1
                    mytable.add_row(j)
            print(mytable)
        return j
                
                
                
    def place_order(self):
        l1=eval(input('select s.no\'s to select an items:'))
        print(' youer selected items are:',l1)
        print('what you selected is...')
        with open('add_data.csv','r') as fo:
            ro=reader(fo)
            l=[]
            m=[]
            for i in ro:
                l.append(i)
            j=[]
            count=0
            num=1
            for j in l:
                if(count==0):
                    j.insert(0,'s.no')
                    count=1
                else:
                    j.insert(0,num)
                    num+=1
                m.append(j)            
            for k in l1:
                for n in m:
                    if(n[0]==k):
                        with open('history.txt','a') as fo:
                            data='you ordered {} of cost {}'.format(n[2],n[4])+'\n'
                            fo.write(data)
                
        self.main_user_page_board()
                            
    def order_history(self):
        with open('history.txt','r') as fo:
            data=fo.read()
            print(data)
        self.main_user_page_board()
    def update_profile(self):
        l=[]
        with open('reg.csv','r',newline='') as cf:
            c=reader(cf)  
            for row in c:
                l.append(row)
        with open('reg.csv','w',newline='') as cf:
            c=writer(cf)
            for i in l:
                if i[2]== self.user_name:
                    continue
                else:
                    c.writerow(i)
            self.fullname=input('enter new name:')
            self.phoneno=int(input('enter new phoneno:'))
            self.emailid=input('enter new emailid:')
            self.address=input('enter new address:')
            self.password=input('enter new password:')
            c.writerow([self.fullname,self.phoneno,self.emailid,self.address,self.password])
        self.main_user_page_board()
            
            
                
        
        
        
            
                    
                    
admin_o=Admin()
user_o=user()
user_o.user_or_admin()
os.remove("history.txt")
```
